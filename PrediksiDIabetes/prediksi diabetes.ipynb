**1. Mengimport library**

```python
import numpy as np
import pandas as pd
import pickle
import streamlit as sl
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

**2. Load dataset**

```python
diabetes_dataset = pd.read_csv('diabetes.csv')
```

```python
diabetes_dataset.head()
```

```python
diabetes_dataset.shape
```

```python
# Memisahkan data dan label
X = diabetes_dataset.drop(columns="Outcome", axis=1)
Y = diabetes_dataset["Outcome"]
```

```python
X
```

```python
Y
```

**3. Standarisasi data**

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standarized_data = scaler.transform(X)
```

```python
standarized_data
```

```python
X = standarized_data
Y = diabetes_dataset["Outcome"]
```

**4. Memisahkan data training dan data testing**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**5. Membuat data latih dengan menggunakan algoritma SVM**

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

**6. Membuat model evaluasi untuk mengukur tingkat akurasi**

```python
X_train_pred = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_pred, Y_train)
```

```python
print("Akurasi data training:", training_data_accuracy)
```

```python
X_test_pred = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_pred, Y_test)
```

```python
print("Akurasi data test:", test_data_accuracy)
```

**7. Membuat model prediksi**

```python
input_data = np.array([6, 148, 72, 35, 0, 33.6, 0.627, 50]).reshape(1, -1)
std_data = scaler.transform(input_data)
```

```python
prediction = classifier.predict(std_data)

if (prediction[0] == 0):
    print("Pasien tidak terkena diabetes :D")
else:
    print("Pasien terkena diabetes T_T")
```

**8. Simpan model yang sudah dilatih**

```python
file_name = "diabetes_model.sav"
pickle.dump(classifier, open(file_name, "wb"))
```

```python
# Simpan scaler
pickle.dump(scaler, open("scaler.sav", "wb"))

# Simpan model
pickle.dump(classifier, open("diabetes_model.sav", "wb"))
```
